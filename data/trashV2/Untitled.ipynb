```python
import os
import xml.etree.ElementTree as ET
#train_img_path = 'train/JPEGImages/'
#train_ann_path = 'train/Annotations/'
test_img_path = 'test/JPEGImages/'
test_ann_path = 'test/Annotations/'
classfile = open('testclasses.csv','w')
annotations_train = open('test_annotations.csv','w')
classes = []
for index,xml in enumerate(os.listdir(test_ann_path)):
    print(xml,str((100*index)/len(os.listdir(test_ann_path)))+"%")
    tree = ET.parse(os.path.join(test_ann_path,xml))
    root = tree.getroot()
    objects = root.findall('object')
    for obj in objects:
        labelname = obj.find('name')
        box = obj.find('bndbox')
        xmin = box.find('xmin')
        ymin = box.find('ymin')
        xmax = box.find('xmax')
        ymax = box.find('ymax')
        if labelname.text not in classes:
            classes.append(labelname.text)
            classfile.write(classes[-1]+","+str(classes.index(labelname.text))+"\n")
        write = os.path.join(test_img_path,xml.split(".")[0]+".jpg")+","+xmin.text+","+ymin.text+","+xmax.text+","+ymax.text+","+labelname.text+"\n"
        annotations_train.write(write)
    annotations_train.flush()
    classfile.flush()
```
